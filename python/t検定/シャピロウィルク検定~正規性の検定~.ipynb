# t検定 step1 正規性の確認
t検定はデータが正規分布である事を前提としている検定である。    

そのため、シャピロ・ウィルク検定によりデータの正規性を検定します。

```python
import numpy as np
from numpy.random import *
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pylab
import statsmodels.api as sm
import pylab
```

```python
#テストデータを作成

#標準正規分布に従う乱数を1000個生成する
normal = randn(1000) 

#比較用の乱数を1000個生成する
comparison = rand(1000)
```

## ヒストグラムで視覚的に確認する

```python
plt.hist(normal,bins=100)
plt.show()
```

```python
plt.hist(comparison,bins=100)
plt.show()
```

## qqプロットで視覚的に確認する

```python
#標準性分布に従う方のqqplot
sm.qqplot(normal, line='45')
pylab.show()
```

```python
#比較用のqqplot
sm.qqplot(comparison, line='45')
pylab.show()
```

### 先人の知恵を借りる(関数を使う)
帰無仮説 : 正規分布である　　　　

対立仮説 : 正規分布ではない

```python
# 出力は(W,p値)
stats.shapiro(normal)

#p値が0.05未満の場合は棄却。今回は0.05以上なので棄却できず、正規分布と判断
```

```python
stats.shapiro(comparison)

#p値が0.05以下の場合は棄却。今回は0.05以下なので棄却。正規分布と判断できない
```
